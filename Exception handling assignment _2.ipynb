Q5. Explain ImportError. What is ModuleNotFoundError?

Ans-
ImportError: ImportError is a base class for exceptions that occur when there are problems during the import of a module. This can happen for various reasons, such as the module not being found, a circular import, or issues with the module's code execution during import. It's a generic exception that covers a wide range of import-related issues

ModuleNotFoundError:
ModuleNotFoundError is a subclass of ImportError and is raised when a module could not be found during import. This more specific exception provides a clearer indication that the issue is specifically related to the module not being found.




```python
''' Example of ImportError'''
try :
    import rohit
except ImportError as e:
    print(e)
```

```python
'''Example of ModuleNotFoundError'''
try:
    import non_existent_module
except ModuleNotFoundError as e:
    print(f"It will show ",e)
```

Q6. List down some best practices for exception handling in python.

Ans-here i have listed  some best practices for exception handling:

1.Specific Exception Handling: 
Catch specific exceptions rather than using a broad Exception catch-all. This allows us  to handle different types of errors differently and provides more accurate error messages.

2.try-except Blocks Sparingly: 
Only use try-except blocks around the specific statements that might raise exceptions. This helps localize the error-handling code to the relevant part of your codebase.

3.finally for Cleanup:
If we need to perform cleanup actions, like closing files or releasing resources, regardless of whether an exception occurred, use a finally block.

4.Don't Suppress Errors: 
we should avoid using empty except blocks that suppress errors without any action. If we encounter an exception, at least log or print relevant information for debugging purposes.

5.Logging:
 We should use a logging library (e.g., logging) to record and trace exceptions. Logging provides a more structured way to collect error information compared to print statements.

6.Raise Exceptions Sparingly: 
Raise exceptions only when necessary and for exceptional cases.we should not  use exceptions for normal flow control.

7.Custom Exception Classes: Create custom exception classes for specific error scenarios in your application. This improves code readability and helps you organize and manage error-handling logic.
